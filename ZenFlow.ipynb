```python
from tensorflow.keras.applications import Xception
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.models import Model
from tensorflow.keras.metrics import Accuracy
import numpy as np
```

```python
xcep= Xception(include_top=False,weights="imagenet",input_shape=[224,224,3])
```

```python
for layer in xcep.layers:
  layer.trainable = False
```

```python
x= Flatten()(xcep.output)
```

```python
prediction = Dense(5,activation='softmax')(x)
```

```python
model =Model(inputs=xcep.input,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
data= ImageDataGenerator(horizontal_flip=True,shear_range=0.2,zoom_range=0.2,rescale=1./255,validation_split=0.3)
```

```python
import os
path = '/content/drive/MyDrive/YogaPoses/'
if os.path.exists(path):
    print("Folders found:", os.listdir(path))
else:
    print("The YogaPoses folder was not found. Check the path again!")
```

```python
training_set=data.flow_from_directory(dataset,target_size=(224,224),batch_size=16,class_mode='categorical',subset='training')
```

```python
testing_set = data.flow_from_directory(dataset,target_size=(224,224),batch_size=16,subset='validation',class_mode='categorical')
```

```python
len(training_set),len(testing_set)
```

```python
model.fit(training_set,validation_data=testing_set,epochs=26,steps_per_epoch=len(training_set),validation_steps=len(testing_set))
```

```python
model.save('xcep_yoga.keras')
```

```python
from tensorflow.keras.models import load_model
model = load_model('xcep_yoga.keras')
```

```python
img= load_img("/content/drive/MyDrive/YogaPoses/Plank/00000000.jpg", target_size=(224,224))
img= image.img_to_array(img)
```

```python
X = np.expand_dims (img, axis=0)
img_data=preprocess_input(X)
img_data.shape
```

```python
pred = model.predict(img_data)
```

```python
P = np.argmax(pred)
columns = ['Downdog', 'Goddess', 'Plank', 'Tree', 'Warrior2']
result = str(columns[P])
```
